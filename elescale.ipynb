```python
import numpy as np
import torch
import matplotlib.pyplot as plt

from hybra.utils import freqtoaud, audtofreq
```

```python
def scale(f):
    return 9.2645 * np.log(1 + f * 0.00437)

def ele_scale(f):
    return 9.2645 * np.log(1 + f * 0.00437 * 2 ) / 2

def inv_scale(f):
    return (np.exp(f / 9.2645) - 1) / 0.00437

def inv_ele_scale(f):
    return (np.exp(f / 9.2645 * 2) - 1) / 0.00437 / 2
```

```python
fs = 16000

plt.figure(figsize=(6, 2))
f = np.linspace(0, fs/2, 1000)
plt.plot(f, scale(f))
plt.ylim(0, 40)

plt.figure(figsize=(6, 2))
f = np.linspace(0, fs/2, 1000)
plt.plot(f, ele_scale(f))
plt.ylim(0, 40)
```
